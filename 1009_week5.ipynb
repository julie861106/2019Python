# map and filter

```python
items = [500, 30, 2000]
c = 30.8594
```

```python
egg = []
for i in items:
    egg.append(i*c)
```

```python
egg
```

```python
def exchange(p):
    return p*c
```

```python
exchange(500)
```

```python
list(map(exchange, items))
```

```python
egg = list(map(exchange, items))
```

```python
egg
```

# lambda 救急函數

```python
list(map(lambda p:p*c, items))
```

# list comprehesion

```python
[p*c for p in items]
```

# filter

```python
grades = [50, 59, 93, 100, 63, 41, 88]
```

```python
def spam(g):
    return g>=60
```

```python
spam(70)
```

```python
list(filter(spam, grades))
```

```python
list(filter(lambda g:g>=60, grades))
```

# print格式化

```python
name = "Julie"
age = 20
```

```python
print("您好, 我是" + name + ", 今年" + str(age) + "歲。")
```

### method01: %    s字串 d數值

```python
egg = "您好, 我是%s, 今年%d歲。"%(name, age)
```

```python
print(egg)
```

### method02: format

```python
egg = "您好, 我是{}, 今年{}歲。".format(name, age)
```

```python
print(egg)
```

### method03: f-string

```python
egg = f"您好, 我是{name}, 今年{age}歲。"
```

```python
print(egg)
```

### 數字照我們的意思顯示

```python
c = 30.8594353
```

```python
print(f"1美金和台幣{c}元")
```

```python
# 取兩位
print(f"1美金和台幣{c:.2f}元")
```

# 更完整print控制

```python
egg = "(✿ﾟ▽ﾟ)"
print(egg)
```

### round1

```python
spam = f"ヽ{egg:20s}ノ"
print(spam)
```

### round2

```python
spam = f"ヽ{egg:>20s}ノ"
print(spam)
```

### round3:置中

```python
spam = f"ヽ{egg:^20s}ノ"
print(spam)
```

### 印星星

```python
level = 5
```

```python
for i in range(level):
    s = "*"*(2*i+1)
    print(f"{s:^21s}")
```

```python
spam = f"ヽ{egg:*>20s}ノ"
print(spam)
```

```python
spam = f"ヽ{egg:*^20s}ノ"
print(spam)
```

```python
n = 512

spam = f"{str(n):0>3s}"
print(spam)
```

```python
n = 87
spam = "000" + str(n)
print(spam)
print(spam[-3:])
```

### zfill不足補0

```python
n = str(87)

spam = n.zfill(3)
print(spam)
```
