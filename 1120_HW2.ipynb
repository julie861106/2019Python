# contour畫圖練習2

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.ion()
plt.figure(figsize=(15,7))

for k in range(10):
    plt.clf()
    plt.subplot(121,polar=True)
    plt.contour(np.random.randn(10,10))
    plt.colorbar()

plt.draw()
```

```python
plt.ion()
plt.figure(figsize=(15,7))

for k in range(10):
    plt.clf()
    plt.subplot(121,polar=True)
    plt.contourf(np.random.randn(10,10))
    plt.colorbar()

plt.draw()
```
