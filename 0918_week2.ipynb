## 日曆小幫手

```python
import calendar

print('歡迎來到日曆小幫手')

while True:
    
    a = input('印出年曆請按1，印出月曆請按2，查詢是否為閏年請按3，若欲跳出請按Q：')
    
    if a == "q" or a == "Q":
        print("下次見～")
        break

    if a == '1':
        y = int(input('印出年曆，請輸入年份：'))
        calendar.prcal(y)
    elif a == '2':
        y = int(input('印出月曆，請輸入年份：'))
        m = int(input('請輸入月份：'))
        calendar.prmonth(y, m)
    elif a == '3':
        y = int(input('查詢是否為閏年，請輸入年份：'))
        print(calendar.isleap(y))
    else:
        print('請輸入1~3')
```

```python
def pipi1(x):
    if x == 'bye':
        print('bye~')
    else:
        print('乖乖')
    
def pipi2(x):
    print('白痴')

print('請問今天要開啟何種模式？')
a = input('1：安慰模式，2：謾罵模式，bye：跳出\n')
x = ''

while x != 'bye':
    if a == '1':
        x = input('>>')
        pipi1(x)
    elif a == '2':
        x = input('>>')
        pipi2(x)
```
