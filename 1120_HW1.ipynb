# contour畫圖練習1

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
data = np.random.randn(800).reshape(10,10,8)
fig,ax = plt.subplots()

def animate(i):
    ax.clear()
    ax.contour(data[:,:,i])

interval = 2
#in seconds 

ani = animation.FuncAnimation(fig,animate,5,interval=interval*1e+3,blit=False)
plt.show()
```

```python
data = np.random.randn(800).reshape(10,10,8)
fig,ax = plt.subplots()

def animate(i):
    ax.clear()
    ax.contourf(data[:,:,i])

interval = 2
#in seconds 

ani = animation.FuncAnimation(fig,animate,5,interval=interval*1e+3,blit=False)
plt.show()
```
